Qualifying h2h

```python
# Ensure constructors_df contains the necessary columns
print("Constructors DataFrame columns:", constructors_df.columns)

# Merge constructors_df to replace constructorId with the constructor name
qualifying_results_2023 = qualifying_results_2023.merge(
    constructors_df[['constructorId', 'name']],
    on='constructorId',
    how='left'
)

# Rename the constructor name column to 'constructor_name'
if 'name_y' in qualifying_results_2023.columns:
    qualifying_results_2023.rename(columns={'name_y': 'constructor_name'}, inplace=True)

# Group by constructor name and round to calculate head-to-head points
def calculate_head_to_head(group):
    group = group.sort_values(by='position')  # Sort by qualifying position
    if len(group) == 2:  # Ensure there are exactly two teammates
        # Assign a head-to-head point to the better performer
        group.iloc[0, group.columns.get_loc('head_to_head_points')] += 1
    return group

# Add a column to track head-to-head points
qualifying_results_2023['head_to_head_points'] = 0

# Apply head-to-head calculation
try:
    qualifying_results_2023 = qualifying_results_2023.groupby(['constructor_name', 'round']).apply(calculate_head_to_head)
except KeyError as e:
    print("Error during groupby operation:", e)
    print("Available columns:", qualifying_results_2023.columns)

# Reset the index to avoid ambiguities
qualifying_results_2023 = qualifying_results_2023.reset_index(drop=True)

# Sum up head-to-head points across the season for each driver
season_head_to_head = (
    qualifying_results_2023.groupby(['constructor_name', 'driverName'])['head_to_head_points']
    .sum()
    .reset_index()
    .sort_values(by=['constructor_name', 'head_to_head_points'], ascending=[True, False])
)

# Display head-to-head comparison
for constructor_name in season_head_to_head['constructor_name'].unique():
    constructor_drivers = season_head_to_head[season_head_to_head['constructor_name'] == constructor_name]
    print(f"Head-to-Head Results for {constructor_name}:\n")
    print(constructor_drivers[['driverName', 'head_to_head_points']].to_string(index=False))
    print("\n" + "="*50 + "\n")
```

Races h2h

```python
# Filter races for the 2023 season
races_2023 = races_df[races_df['year'] == 2023]
race_ids_2023 = races_2023['raceId']

# Filter results for the 2023 season
results_2023 = results_df[results_df['raceId'].isin(race_ids_2023)]

# Merge race results with constructors_df to replace constructorId with the constructor name
results_with_constructors = results_2023.merge(
    constructors_df[['constructorId', 'name']],
    on='constructorId',
    how='left'
)

# Rename the constructor name column for clarity
results_with_constructors.rename(columns={'name': 'constructor_name'}, inplace=True)

# Merge with drivers_df to get driver names
results_with_constructors = results_with_constructors.merge(
    drivers_df[['driverId', 'forename', 'surname']],
    on='driverId',
    how='left'
)

# Create a full driver name column
results_with_constructors['driverName'] = results_with_constructors['forename'] + ' ' + results_with_constructors['surname']

# Add a column to track head-to-head points
results_with_constructors['head_to_head_points'] = 0

# Define a function to calculate head-to-head points
def calculate_race_head_to_head(group):
    group = group.sort_values(by='positionOrder')  # Sort by race finishing position
    if len(group) == 2:  # Ensure there are exactly two teammates
        # Assign a head-to-head point to the better performer
        group.iloc[0, group.columns.get_loc('head_to_head_points')] += 1
    return group

# Apply head-to-head calculation grouped by constructor name and race
results_with_constructors = results_with_constructors.groupby(['constructor_name', 'raceId']).apply(calculate_race_head_to_head)

# Reset the index to avoid ambiguities
results_with_constructors = results_with_constructors.reset_index(drop=True)

# Sum up head-to-head points across the season for each driver
race_head_to_head = (
    results_with_constructors.groupby(['constructor_name', 'driverName'])['head_to_head_points']
    .sum()
    .reset_index()
    .sort_values(by=['constructor_name', 'head_to_head_points'], ascending=[True, False])
)

# Display head-to-head comparison for race results in the 2023 season
for constructor_name in race_head_to_head['constructor_name'].unique():
    constructor_drivers = race_head_to_head[race_head_to_head['constructor_name'] == constructor_name]
    print(f"Head-to-Head Results for {constructor_name} (Race Results - 2023):\n")
    print(constructor_drivers[['driverName', 'head_to_head_points']].to_string(index=False))
    print("\n" + "="*50 + "\n")
```

```python
# Filter races for the 2023 season
races_2023 = races_df[races_df['year'] == 2023]
race_ids_2023 = races_2023['raceId']

# Filter results for the 2023 season
results_2023 = results_df[results_df['raceId'].isin(race_ids_2023)]

# Merge race results with constructors_df to get constructor names
results_with_constructors = results_2023.merge(
    constructors_df[['constructorId', 'name']],
    on='constructorId',
    how='left'
)

# Rename constructor name column
results_with_constructors.rename(columns={'name': 'constructor_name'}, inplace=True)

# Merge with drivers_df to get driver names
results_with_constructors = results_with_constructors.merge(
    drivers_df[['driverId', 'forename', 'surname']],
    on='driverId',
    how='left'
)

# Create a full driver name column
results_with_constructors['driverName'] = results_with_constructors['forename'] + ' ' + results_with_constructors['surname']

# Aggregate total points by constructor and driver
driver_points = results_with_constructors.groupby(['constructor_name', 'driverName'])['points'].sum().reset_index()

# Aggregate total points by constructor
constructor_points = results_with_constructors.groupby(['constructor_name'])['points'].sum().reset_index()
constructor_points.rename(columns={'points': 'constructor_total_points'}, inplace=True)

# Merge the driver and constructor points to calculate percentages
driver_points = driver_points.merge(constructor_points, on='constructor_name')
driver_points['percentage'] = (driver_points['points'] / driver_points['constructor_total_points']) * 100

# Plot pie charts for each constructor
constructors = driver_points['constructor_name'].unique()

for constructor in constructors:
    constructor_data = driver_points[driver_points['constructor_name'] == constructor]
    labels = constructor_data.apply(
        lambda row: f"{row['driverName']} ({row['points']} pts)", axis=1
    )
    sizes = constructor_data['percentage']

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f"{constructor} - Points Distribution (2023)")
    plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
    plt.show()
```

Analysis of who is the best driver of 2023:

formula:
ARR = Adjusted Race Results
CQP = Car Qualifying Performance
Driver Skill (DS) = ARR/CQP

for eg. 
Driver finished 3rd 
Car was the 4th best car on average based on qualifying results for that weekend

- ARR = 20 / 3 = 6.67
- CQP = 4
- DS = 6.67 / 4 = 1.67

if DS > 1, means the driver has outperformed the car
if DS ~ 1, means the driver is performing at the car level
if DS < 1, means the driver is underperforming

things to note: 
the way to calculate the qualifying results is the average of all the qualifying results of the team's 2 drivers that weekend. 

for eg. max verstappen completed q3 that weekend and perez only managed q2, the average is between the q1, q2, q3 of Verstappen and the q1 and q2 timings of perez. average timing = avg(q1Verstappen,q2Verstappen,q3Verstappen,q1Perez,q2Perez)


```python
import numpy as np
import pandas as pd

# Filter data for 2023 season
races_2023 = races_df[races_df['year'] == 2023]
race_ids_2023 = races_2023['raceId']
qualifying_2023 = qualifying_df[qualifying_df['raceId'].isin(race_ids_2023)]

# Replace '\N' with NaN and convert qualifying times to seconds
qualifying_2023 = qualifying_2023.replace("\\N", np.nan)

# Function to convert time strings (e.g., '1:31.295') to seconds
def time_to_seconds(time_str):
    if pd.isnull(time_str):
        return None
    try:
        minutes, seconds = map(float, time_str.split(":"))
        return minutes * 60 + seconds
    except:
        return None

# Apply conversion function to qualifying times
for col in ['q1', 'q2', 'q3']:
    qualifying_2023[col] = qualifying_2023[col].apply(time_to_seconds)

# Calculate the fastest time for each driver
qualifying_2023['fastest_time'] = qualifying_2023[['q1', 'q2', 'q3']].min(axis=1)

# Merge with constructors_df and races_df for additional context
qualifying_2023 = qualifying_2023.merge(
    constructors_df[['constructorId', 'name']],
    on='constructorId',
    how='left'
)
qualifying_2023 = qualifying_2023.merge(
    races_2023[['raceId', 'name']],
    on='raceId',
    how='left'
)

# Group by race and constructor to calculate average fastest time for each team
constructor_avg_times = (
    qualifying_2023.groupby(['raceId', 'constructorId'])['fastest_time']
    .mean()
    .reset_index()
)

# Rank constructors within each race based on their average fastest times
constructor_avg_times['CQP'] = constructor_avg_times.groupby('raceId')['fastest_time'].rank(method='dense',ascending=False)

# Merge CQP with the race results to calculate ARR and DS
results_with_cqp = results_df[results_df['raceId'].isin(race_ids_2023)].merge(
    constructor_avg_times[['raceId', 'constructorId', 'CQP']],
    on=['raceId', 'constructorId'],
    how='left'
)

# Calculate ARR = 20 / finishing position
results_with_cqp['ARR'] = 20 / results_with_cqp['positionOrder']

# Calculate Driver Skill (DS) = ARR / CQP
results_with_cqp['DS'] = results_with_cqp['ARR'] / results_with_cqp['CQP']

# Merge with driver and constructor names for better context
results_with_cqp = results_with_cqp.merge(
    drivers_df[['driverId', 'forename', 'surname']],
    on='driverId',
    how='left'
).merge(
    constructors_df[['constructorId', 'name']],
    on='constructorId',
    how='left'
)

# Add race name for final display
results_with_cqp = results_with_cqp.merge(
    races_2023[['raceId', 'name']],
    on='raceId',
    how='left'
).rename(columns={'name_x': 'constructor', 'name_y': 'race_name'})

# Select and sort final columns
final_results = results_with_cqp[['raceId', 'race_name', 'forename', 'surname', 'constructor', 'CQP', 'ARR', 'DS']]
final_results = final_results.sort_values(by=['raceId', 'DS'], ascending=[True, False])

# Reset index for clarity
final_results_df = final_results.reset_index(drop=True)

# Display the DataFrame
print("\nDriver Performance Analysis (2023):")
(final_results_df)
```

```python
# Ensure the data is sorted by raceId to maintain chronological order
final_results_df = final_results_df.sort_values(by=['raceId'])

# Define teammate pairings as a dictionary
team_pairings = {
    "Red Bull": ["Verstappen", "Pérez"],
    "Ferrari": ["Leclerc", "Sainz"],
    "Mercedes": ["Hamilton", "Russell"],
    "Aston Martin": ["Alonso", "Stroll"],
    "Alpine": ["Ocon", "Gasly"],
    "McLaren": ["Norris", "Piastri"],
    "Haas F1 Team": ["Magnussen", "Hülkenberg"],
    "Alfa Romeo": ["Bottas", "Zhou"],
    "Williams": ["Albon", "Sargeant"]
}

# Plot each teammate pairing
for team, drivers in team_pairings.items():
    plt.figure(figsize=(12, 6))

    # Filter data for the two teammates
    pairing_data = final_results_df[final_results_df['surname'].isin(drivers)]
    pivot_table = pairing_data.pivot(index='raceId', columns='surname', values='DS')

    # Plot each driver's DS line across the season
    for driver in drivers:
        if driver in pivot_table.columns:
            plt.plot(pivot_table.index, pivot_table[driver], label=driver, marker='o')

    # Adding labels and title
    plt.xlabel('Race ID (Chronological Order)', fontsize=12)
    plt.ylabel('DS (Driver Score)', fontsize=12)
    plt.title(f"{team} Teammates' DS Across the Season (2023)", fontsize=14)
    plt.legend(title='Driver', loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
# Rank drivers by average DS and sort in descending order
average_ds = final_results_df.groupby('surname')['DS'].mean().sort_values(ascending=False)

# Display the ranked drivers
print("Average DS Rankings (Descending):")
print(average_ds)
```
